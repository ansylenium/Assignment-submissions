```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
X.shape
```

```python
# Fitting PCA 
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
```

```python
X.shape
```

```python
data = pd.concat([pd.DataFrame(X, columns=['x','y','z']), pd.DataFrame(y, columns=['target'])], axis=1)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(data.x, data.y, data.z, c=data.target.map({0:'r', 1:'b', 2:'g'}))
```
